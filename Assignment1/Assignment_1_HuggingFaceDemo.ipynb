```python
from transformers import pipeline

# Load a text generation pipeline
generator = pipeline(
    "text-generation",
    model="gpt2"
)

# Input prompt
prompt = "Space Technology is emerging day by day as there is"

# Generate text
output = generator(
    prompt,
    max_length=60,
    num_return_sequences=1
)

print(output[0]["generated_text"])
```
